# StandardScaler, MinMaxScaler

## Original Data

```python
from sklearn.datasets import load_iris
import pandas as pd

iris = load_iris()
iris_data = iris.data
df = pd.DataFrame(data=iris_data, columns=iris.feature_names)
df.describe()
```

```python
import seaborn as sns

sns.pairplot(df)
```

## StandardScaler



* mean: 0, std: 1

```python
from sklearn.preprocessing import StandardScaler

st_scaler = StandardScaler()
st_scaler.fit(df)
st_scaled_array = st_scaler.transform(df)

standard_scaled_df = pd.DataFrame(data=st_scaled_array, columns=iris.feature_names)
standard_scaled_df.describe()
```

```python
sns.pairplot(standard_scaled_df)
```

## MinMaxScaler

- min = 0, max = 1

```python
from sklearn.preprocessing import MinMaxScaler

mm_scaler = MinMaxScaler()
mm_scaler.fit(df)
mm_scaled_array = mm_scaler.transform(df)

mm_scaled_df = pd.DataFrame(data=mm_scaled_array, columns=iris.feature_names)
mm_scaled_df.describe()
```

```python
sns.pairplot(mm_scaled_df)
```
